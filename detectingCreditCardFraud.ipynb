```python
import pandas as pd  # Import pandas library for data manipulation and analysis

df = pd.read_csv('creditcard.csv')  # Load the dataset from a CSV file into a DataFrame

df  # Display the full DataFrame (for quick inspection in the notebook)
```

```python
#Tells us how many are fraud based on 1(Legit) or 0(Fraud) value in "Class" parameter
df['Class'].value_counts()
```

```python
# Preprocessing

from sklearn.preprocessing import RobustScaler  # Import scaler that is resistant to outliers

# Create a copy of the original dataset to avoid modifying raw data
new_df = df.copy()  

# Scale the 'Amount' column using RobustScaler
# RobustScaler uses median and interquartile range, making it less sensitive to extreme outliers
new_df['Amount'] = RobustScaler().fit_transform(
    new_df['Amount'].to_numpy().reshape(-1, 1)
)

# Normalize the 'Time' column using min-max scaling
time = new_df['Time']
new_df['Time'] = (time - time.min()) / (time.max() - time.min())

new_df  # Display the preprocessed dataset
```

```python
# Shuffle all rows in the dataset
# frac=1 means use 100% of the data
# random_state=1 ensures reproducibility (same shuffle every time)
new_df = new_df.sample(frac=1, random_state=1)  

new_df  # Display the shuffled dataset
```

```python
# Split dataset into training, testing, and validation sets
train, test, val = new_df[:240000], new_df[240000:262000], new_df[262000:]

# Check class distribution in each split
train['Class'].value_counts(), test['Class'].value_counts(),val['Class'].value_counts()
```

```python
# Convert DataFrames to NumPy arrays
train_np = train.to_numpy()
test_np = test.to_numpy()
val_np = val.to_numpy()

# Display the shape of each dataset
train_np.shape, test_np.shape, val_np.shape
```

```python
#This code splits the dataset into features (x) and target variable (y) 
# for training, testing, and validation sets. The last column is assumed 
# to be the target variable ('Class'), while all preceding columns are 
# features. The shapes of the resulting arrays are also displayed to 
# confirm the correct splitting.

#Right now the data is in a big 2D array

# Select all rows (:) and all columns except the last (:-1)
# This extracts the feature matrix used as model input
# So now, in the x_train variable, we have all the features (V1, V2, ..., Amount, Time) 
# but not the target variable (Class) which indicates whether a transaction is fraudulent or not.
x_train = train_np[:, :-1]

# Select all rows (:) and only the last column (-1)
# This extracts the target labels (0 = legitimate, 1 = fraud)
#So now, in the y_train variable, we have the target variable (Class) which indicates 
# whether a transaction is fraudulent (1) or legitimate (0) as a 1D vector
y_train = train_np[:, -1]

# Repeat the same separation for test and validation sets
x_test = test_np[:, :-1]
y_test = test_np[:, -1]

x_val = val_np[:, :-1]
y_val = val_np[:, -1]

# Display shapes to verify correctness
# X should be 2D: (number_of_samples, number_of_features)
# y should be 1D: (number_of_samples,)
x_train.shape, y_train.shape, x_test.shape, y_test.shape, x_val.shape, y_val.shape
```

```python
# Import the Logistic Regression classifier.
# Logistic Regression is a linear model used for binary classification.
# It estimates the probability that a transaction belongs to class 1 (fraud).
from sklearn.linear_model import LogisticRegression


# Initialize the Logistic Regression model using default hyperparameters.
# This creates the model object but does not train it yet.
logistic_model = LogisticRegression()


# Train (fit) the model using the training feature matrix (x_train)
# and corresponding labels (y_train).
# During training, the model learns a weight for each feature that
# helps separate fraudulent from legitimate transactions.
logistic_model.fit(x_train, y_train)


# Evaluate the model on the training data.
# The .score() method returns accuracy:
# (number of correct predictions) / (total number of predictions).
# Note: Accuracy alone is not a reliable metric for imbalanced fraud data.
logistic_model.score(x_train, y_train)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_val, logistic_model.predict(x_val), target_names=['Not Fraud', 'Fraud']))
#Recall is Out of all of the actual fraud cases, how many did you correctly flag as fraud
#Precision is Out of all of the cases you said were fraud, were actually fraud(How many legitimate transactions did you flag)
#Accuracy score is 100% only becasue there is so many not fraud cases compared to fraud but out of the total amount of fraud cases, we only flagged half of them which is not good
```

```python
# Import required components from TensorFlow Keras.
# Sequential is used to build a feed-forward neural network layer by layer.
# InputLayer defines the input shape.
# Dense defines fully connected layers.
# BatchNormalization stabilizes and speeds up training.
# ModelCheckpoint allows saving the best model during training.
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense, BatchNormalization
from tensorflow.keras.callbacks import ModelCheckpoint


# Create a sequential neural network model.
# Layers will be added in order from input to output.
shallow_nn = Sequential()


# Define the input layer.
# The input shape equals the number of features in x_train.
# x_train.shape[1] represents the number of input variables per transaction.
shallow_nn.add(InputLayer((x_train.shape[1],)))


# Add a hidden dense layer with 2 neurons.
# 'relu' activation introduces non-linearity.
# This allows the model to learn more complex decision boundaries
# than linear models like Logistic Regression.
shallow_nn.add(Dense(2, activation='relu'))


# Apply Batch Normalization.
# This normalizes layer outputs during training,
# which can improve stability and convergence speed.
shallow_nn.add(BatchNormalization())


# Add the output layer with 1 neuron.
# 'sigmoid' activation outputs a probability between 0 and 1,
# representing the probability that a transaction is fraudulent.
shallow_nn.add(Dense(1, activation='sigmoid'))


# Set up model checkpointing.
# The best-performing model (based on validation loss during training)
# will be saved to the file 'shallow_nn.keras'.
checkpoint = ModelCheckpoint('shallow_nn.keras', save_best_only=True)


# Compile the model.
# 'adam' is an adaptive gradient-based optimizer.
# 'binary_crossentropy' is the appropriate loss function
# for binary classification problems.
# 'accuracy' is included as a training metric to monitor performance.
shallow_nn.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
# Display a summary of the neural network architecture.
# This shows each layer, its output shape, and the number of trainable parameters.
# It allows us to verify that the model structure matches our expectations.
shallow_nn.summary()
```

```python
# Train the neural network on the training data.
# x_train contains the input features.
# y_train contains the corresponding fraud labels (0 or 1).
# validation_data is used to evaluate performance after each epoch
# to monitor generalization and detect overfitting.
# epochs=5 means the model will pass through the entire training dataset 5 times.
# callbacks=checkpoint enables saving the best model based on validation performance.
shallow_nn.fit(x_train, y_train, validation_data=(x_val,y_val), epochs=5, callbacks=checkpoint)
```

```python
# Define a helper function to generate class predictions from a neural network model.
# model: the trained neural network.
# x: input features to make predictions on.
# The model outputs probabilities (between 0 and 1) because of the sigmoid activation.
# We convert these probabilities to binary labels:
# - Greater than 0.5 → 1 (fraud)
# - Less than or equal to 0.5 → 0 (legitimate)

# model.predict(x) produces probabilities for each sample in x.
# .flatten() converts the output from shape (num_samples, 1) to (num_samples,).
# > 0.5 applies a threshold to classify each sample as fraud (1) or not (0).
# .astype(int) converts the boolean array to integer labels.
def neural_net_predictions(model, x):
    return (shallow_nn.predict(x).flatten() > 0.5).astype(int)

# Generate predictions on the validation set.
# This returns a 1D array of 0s and 1s corresponding to predicted fraud labels.
neural_net_predictions(shallow_nn, x_val)
```

```python
# Compares predicted labels from the neural network with the true labels.
# Computes precision, recall, and F1-score for both classes.
# Prints a formatted report showing how well the model performs on the validation set.
print(classification_report(y_val, neural_net_predictions(shallow_nn, x_val), target_names=['Not Fraud', 'Fraud']))
```

```python
# Import the Random Forest classifier from scikit-learn.
# Random Forest is an ensemble model that builds multiple decision trees
# and averages their predictions for more robust and accurate classification.
from sklearn.ensemble import RandomForestClassifier


# Initialize the Random Forest model.
# max_depth=2 limits each tree to 2 levels, preventing overfitting on this dataset.
# n_jobs=-1 allows using all CPU cores for faster training.
rf = RandomForestClassifier(max_depth=2, n_jobs=-1)


# Train the Random Forest on the training data.
# x_train: input features
# y_train: target labels (0 = Not Fraud, 1 = Fraud)
rf.fit(x_train, y_train)

print(classification_report(y_val, rf.predict(x_val), target_names=['Not Fraud', 'Fraud']))
```

```python
# Import the Gradient Boosting Classifier from scikit-learn.
# Gradient Boosting builds an ensemble of weak learners (usually shallow trees)
# sequentially, where each new tree tries to correct the errors of the previous ones.
from sklearn.ensemble import GradientBoostingClassifier


# Initialize the Gradient Boosting model.
# n_estimators=50: number of trees to build in sequence.
# learning_rate=1.0: step size at each iteration for weight updates.
# max_depth=1: each tree is a "stump" (very shallow) to prevent overfitting.
# random_state=0 ensures reproducible results.
gbc = GradientBoostingClassifier(n_estimators=50, learning_rate=1.0, max_depth=1, random_state=0)
gbc.fit(x_train,y_train)

print(classification_report(y_val, gbc.predict(x_val), target_names=['Not Fraud', 'Fraud']))
```

```python
# Since the dataset is imbalanced, we want to analyze 
# the distribution of legitimate vs fraudulent transactions.

# Separate the dataset into legitimate and fraudulent transactions for analysis.
# new_df.query('Class == 0') selects all rows where Class is 0 (Not Fraud).
# new_df.query('Class == 1') selects all rows where Class is 1 (Fraud).
not_frauds = new_df.query('Class == 0')
frauds = new_df.query('Class == 1')


# Count the number of samples in each subset.
# This confirms the severe class imbalance:
# - not_frauds will contain the vast majority of transactions.
# - frauds will contain only a small fraction.
not_frauds['Class'].value_counts(), frauds['Class'].value_counts()
```

```python
# Create a balanced dataset by undersampling the majority class (Not Fraud).
# - Take all fraudulent transactions (frauds).
# - Randomly sample an equal number of legitimate transactions (not_frauds) to match the number of frauds.
# - pd.concat combines the two subsets into a single balanced DataFrame.
# random_state=1 ensures reproducibility of the sampling.
balanced_df = pd.concat([
    frauds,
    not_frauds.sample(len(frauds), random_state=1)
])


# Count the number of transactions in each class to verify balancing.
# Both classes should now have the same number of samples.
balanced_df['Class'].value_counts()
```

```python
# Shuffle the balanced dataset randomly.
# - frac=1 means use 100% of the rows.
# - random_state=1 ensures the shuffle is reproducible.
# Shuffling prevents any order-based patterns from affecting training.
balanced_df = balanced_df.sample(frac=1, random_state=1)


# Display the shuffled balanced dataset.
# Now the rows of fraud and non-fraud transactions are mixed.
balanced_df
```

```python
# Convert the balanced DataFrame into a NumPy array for easier manipulation.
# This allows for efficient indexing and prepares the data for model training.
balanced_df_np = balanced_df.to_numpy()


# Split the balanced dataset into features (X) and target labels (y) for training, testing, and validation.
# - x_train_b, y_train_b: first 700 samples
# - x_test_b, y_test_b: next 142 samples (700:842)
# - x_val_b, y_val_b: remaining samples (842 onward)
# .astype(int) ensures the target labels are integers (0 or 1)
x_train_b, y_train_b = balanced_df_np[:700, :-1], balanced_df_np[:700, -1].astype(int)
x_test_b, y_test_b = balanced_df_np[700:842, :-1], balanced_df_np[700:842, -1].astype(int)
x_val_b, y_val_b = balanced_df_np[842:, :-1], balanced_df_np[842:, -1].astype(int)


# Display the shapes of all splits to verify correctness.
# Expected shapes:
# - X matrices: (num_samples, num_features)
# - y vectors: (num_samples,)
x_train_b.shape, y_train_b.shape, x_test_b.shape, y_test_b.shape, x_val_b.shape, y_val_b.shape


# Dataset split: 700 training samples, 142 testing, 142 validation.
# Each sample has 30 features, consistent across all splits. (V1-28 + Amount + Time)
```

```python
# Check the class distribution in the training, testing, and validation splits.
# Convert the label arrays (y_train_b, y_test_b, y_val_b) into pandas Series to use value_counts().
# This helps verify that the balanced dataset maintained an equal number of fraud (1) and non-fraud (0) samples in each split.
import pandas as pd

pd.Series(y_train_b).value_counts(), \
pd.Series(y_test_b).value_counts(), \
pd.Series(y_val_b).value_counts()
```

```python
# Train a Logistic Regression model on the balanced dataset.
# Logistic Regression is a simple linear model for binary classification.
# Using the balanced dataset ensures the model sees roughly equal numbers
# of fraud (1) and non-fraud (0) transactions during training.
from sklearn.linear_model import LogisticRegression

logistic_model_b = LogisticRegression()


# Fit the model to the training features and labels.
# x_train_b: features of the balanced training set
# y_train_b: corresponding labels (0 = Not Fraud, 1 = Fraud)
logistic_model_b.fit(x_train_b, y_train_b)


# Evaluate the trained model on the balanced validation set.
# classification_report provides precision, recall, F1-score, and support for each class.
# target_names labels the classes for clarity.
from sklearn.metrics import classification_report

print(classification_report(
    y_val_b,
    logistic_model_b.predict(x_val_b),
    target_names=['Not Fraud', 'Fraud']
))
```

```python
# Build a shallow neural network for binary classification on the balanced dataset.
# Sequential allows adding layers in order.
# InputLayer defines the input shape based on the number of features (30 in this case).
# Dense layers define fully connected layers.
# BatchNormalization normalizes activations to stabilize and speed up training.
# The final Dense layer uses sigmoid activation to output a probability of fraud (1).
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense, BatchNormalization
from tensorflow.keras.callbacks import ModelCheckpoint

shallow_nn_b = Sequential()

# Input layer
shallow_nn_b.add(InputLayer((x_train_b.shape[1],)))

# Hidden layer with 2 neurons and ReLU activation
# Relu (Rectified Linear Unit) allows the model to learn complex patterns between features and the target variable.
shallow_nn_b.add(Dense(2, activation='relu'))

# Normalize the hidden layer outputs
shallow_nn_b.add(BatchNormalization())

# Output layer with sigmoid activation for binary classification
shallow_nn_b.add(Dense(1, activation='sigmoid'))

# Save the best version of the model during training
# based on validation loss
checkpoint = ModelCheckpoint('shallow_nn_b.keras', save_best_only=True)

# Compile the model
# Adam optimizer, binary cross-entropy loss (for 0/1 classification)
# Accuracy metric (though precision/recall is more meaningful for fraud detection)
shallow_nn_b.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Train the model on the balanced dataset
# x_train_b, y_train_b: balanced training features and labels
# validation_data: evaluate performance on balanced validation set after each epoch
# epochs=40: train for 40 full passes over the training data
# callbacks=checkpoint: save the best-performing model
shallow_nn_b.fit(
    x_train_b,
    y_train_b,
    validation_data=(x_val_b, y_val_b),
    epochs=40,
    callbacks=checkpoint
)
```

```python
# Evaluate the shallow neural network trained on the balanced dataset.
# y_val_b: true labels for the validation set
# neural_net_predictions(shallow_nn_b, x_val_b): predicted labels from the trained network
# target_names: labels for clarity (0 = 'Not Fraud', 1 = 'Fraud')
# classification_report shows precision, recall, F1-score, and support for each class.

print(classification_report(
    y_val_b,
    neural_net_predictions(shallow_nn_b, x_val_b),
    target_names=['Not Fraud', 'Fraud']
))
```
